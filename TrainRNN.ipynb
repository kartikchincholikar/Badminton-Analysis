## Restoring Model

```python
Train = False
try:
    sess = tf.InteractiveSession()
    saver = tf.train.import_meta_graph(directory+"/"+model_params+".meta")
    saver.restore(sess,tf.train.latest_checkpoint(directory+"/"))
    graph = tf.get_default_graph()
    
    input_to_dynamic = graph.get_tensor_by_name('inputs:0')
    predicted = graph.get_tensor_by_name('labels:0')
    rally_lengths = graph.get_tensor_by_name('rally_length:0')
    softmax_output = graph.get_tensor_by_name('softmax_output:0')
    sigmoid_output = graph.get_tensor_by_name('sigmoid_output:0')
    
except OSError:
    Train = True
    
```

## Helper Functions`

```python
def batch_iter(data,lengths,batch_size, num_epochs, shuffle=True):
    """
    Generates a batch iterator for a dataset.
    """
    original_data = np.array(data)
    
    data = original_data[:,:-1]
    labels = original_data[:,1:]
    sentences_lengths = np.array(lengths)
    
    data_size = len(data)
    num_batches_per_epoch = int((len(data) - 1) / batch_size) + 1
    for epoch in range(num_epochs):
        # Shuffle the data at each epoch
        if shuffle:
            shuffle_indices = np.random.permutation(np.arange(data_size))
            shuffled_data = data[shuffle_indices]
            shuffled_labels = labels[shuffle_indices]
            shuffled_lengths = sentences_lengths[shuffle_indices]
        else:
            shuffled_data = data
            shuffled_labels = labels
            shuffled_lengths = sentences_lengths
            
        for batch_num in range(num_batches_per_epoch):
            start_index = batch_num * batch_size
            end_index = min((batch_num + 1) * batch_size, data_size)
            yield (shuffled_data[start_index:end_index],shuffled_labels[start_index:end_index],shuffled_lengths[start_index:end_index],epoch)
```

```python
def predict(sentences_2predict,text_processor,vocabulary,sess):
    num_sentences = len(sentences_2predict)
    lengths = [len(x.split()) for x in sentences_2predict]
    int_sentences_p = np.array(list(text_processor.transform(sentences_2predict)))
    int_sentences_p = int_sentences_p[:,:-1]

    feed_dict = {input_to_dynamic:int_sentences_p,rally_lengths:lengths}

    prediction = sess.run([softmax_output],feed_dict=feed_dict)[0]
    
    return [(vocabulary[np.argmax(prediction[x,lengths[x]-1,:])],prediction[x,lengths[x]-1,:]) for x in range(num_sentences)]


def validate(sess):
    val_batches = batch_iter(test_int,test_len,1,1)
    correct_stroke_lists = []
    counter = 0
    perfectly_correct=0
    for batch in val_batches:
        inputs = batch[0]
        targets = batch[1]
        stroke_list = ' '
        for char,next_char in zip(inputs[0,:],targets[0,:]):
            if char ==0 or char >32:
                break
            stroke_list +=vocabulary[char]+" "
            #print(stroke_list)
            #print(vocabulary[next_char])
            prediction = predict([stroke_list],text_processor,vocabulary,sess)

            #print("Prediction is {}".format(prediction[0][0]))

            #print("______________________________________________")
            if prediction[0][0]==vocabulary[next_char]:
                perfectly_correct +=1
                correct_stroke_lists.append(stroke_list)

            counter+=1
    return (perfectly_correct/counter,correct_stroke_lists)
```

## Training Model

```python
if Train:
    #convert sentence to ints
    int_sentences = text_processor.transform(sentences)
    len_sentences = [len(x.split()) for x in sentences]

    train_int,test_int,train_len,test_len = train_test_split(list(int_sentences),len_sentences,test_size = 0.2)

    tf.reset_default_graph()
    input_to_dynamic = tf.placeholder(tf.int32,[None,39],name="inputs")
    labels = tf.placeholder(tf.int32,[None,39],name="labels")
    rally_lengths = tf.placeholder("float",[None],name="rally_length")


    #INPUTS
    embedded_chars = tf.nn.embedding_lookup(encodings_sigmoid,input_to_dynamic)

    #TARGETS
    #for softmax targets:
    one_hot_labels = tf.one_hot(labels,41)
    #for sigmoid_targets
    distributed_labels = tf.nn.embedding_lookup(encodings_sigmoid,labels)

   
    cell = tf.nn.rnn_cell.BasicRNNCell(num_Hidden_Neurons )
    output_states,final_state = tf.nn.dynamic_rnn(cell,embedded_chars,sequence_length=rally_lengths,dtype=tf.float32)
     #RNN GRAPH CREATE FOR SOFTMAX
    pre_softmax = tf.contrib.layers.fully_connected(output_states
                                                     ,num_outputs=41
                                                    ,activation_fn=None,weights_initializer = tf.initializers.truncated_normal)


    softmax_output = tf.nn.softmax(pre_softmax,name="softmax_output")


    softmax_cost = tf.losses.softmax_cross_entropy(onehot_labels=one_hot_labels,logits=pre_softmax)

    softmax_optimizer = tf.train.AdamOptimizer().minimize(softmax_cost)
    #RNN GRAPH FOR SIGMOID
    pre_sigmoid = tf.contrib.layers.fully_connected(output_states
                                                     ,num_outputs=7
                                                    ,activation_fn=None,weights_initializer = tf.initializers.truncated_normal)
    
    sigmoid_output = tf.nn.sigmoid(pre_sigmoid,name="sigmoid_output")
    
    sigmoid_cost = tf.losses.sigmoid_cross_entropy(tf.reshape(distributed_labels,[-1,7]),tf.reshape(sigmoid_output,[-1,7]))
    sigmoid_optimizer = tf.train.AdamOptimizer().minimize(sigmoid_cost)
    

    #we do mini batching
    batches = batch_iter(train_int,train_len,mini_batch_size,epochs)

    sess = tf.InteractiveSession()
    sess.run(tf.global_variables_initializer())

    prev_epoch = -1
    for batch in batches:
        input_data = batch[0]
        y_labels = batch[1]
        lengths = batch[2]

        epoch = batch[3]


        _,cost = sess.run([sigmoid_optimizer,sigmoid_cost],feed_dict={input_to_dynamic:input_data,#softmax_optimizer,softmax_cost
                                                               labels:y_labels,
                                                             rally_lengths:lengths})

        if epoch!=prev_epoch and epoch%200==0:
            print("Training cost Epoch {} is : {}".format(epoch,cost))
        if epoch!=prev_epoch and epoch%500==0:
            validation_accuracy = validate(sess)
            print("**Validation accuracy Epoch {} is {}".format(epoch,validation_accuracy[0]))
        prev_epoch = epoch

    #Saving the model   
    saver = tf.train.Saver()
    saver.save(sess,model_path)

```
